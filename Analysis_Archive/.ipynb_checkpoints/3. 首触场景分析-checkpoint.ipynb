```python
# 三大垂媒平台用户特征 - 车系分布
'''
select 
    c.level_4_tp_id,
    series_id,
    count(distinct a.mobile) as vol
from
(
    select mobile, last_fir_contact_date_brand, touchpoint_id
    from
    (
        select mobile, last_fir_contact_date_brand, touchpoint_id,
        row_number() over(partition by mobile order by last_fir_contact_date_brand desc) as rank_
        from marketing_modeling.cdm_customer_touchpoints_profile_a
        where brand = "MG"
    ) a
    where rank_ = 1
) a
left join
(
    select * from marketing_modeling.dw_oppor_behavior where brand_id = 121
) b
on a.mobile = b.phone
left join marketing_modeling.cdm_touchpoints_id_system c
on a.touchpoint_id = c.touchpoint_id
where a.last_fir_contact_date_brand <= b.behavior_time
group by
c.level_4_tp_id,
series_id
'''
```

```python
# 三大垂媒平台用户特征 - 年龄分布
'''
select 
    c.level_4_tp_id,
    case when x_age between 18 and 20 then '<20'
    when x_age <= 25 then '20-25'
    when x_age <= 30 then '25-30'
    when x_age <= 35 then '30-35'
    when x_age <= 40 then '35-40'
    when x_age <= 45 then '40-45'
    when x_age <= 80 then '>45'
    else 'others' end as x_age,
    count(distinct a.mobile) as vol
from
(
    select mobile, last_fir_contact_date_brand, touchpoint_id
    from
    (
        select mobile, last_fir_contact_date_brand, touchpoint_id,
        row_number() over(partition by mobile order by last_fir_contact_date_brand desc) as rank_
        from marketing_modeling.cdm_customer_touchpoints_profile_a
        where brand = "MG"
    ) a
    where rank_ = 1
) a
left join
(
    select * from marketing_modeling.dw_core_image_basic_feature
) b
on a.mobile = b.phone
left join marketing_modeling.cdm_touchpoints_id_system c
on a.touchpoint_id = c.touchpoint_id
group by
c.level_4_tp_id,
case when x_age between 18 and 20 then '<20'
when x_age <= 25 then '20-25'
when x_age <= 30 then '25-30'
when x_age <= 35 then '30-35'
when x_age <= 40 then '35-40'
when x_age <= 45 then '40-45'
when x_age <= 80 then '>45'
else 'others' end
'''
```

```python
# 三大垂媒平台用户特征 - 成交周期
'''
select
    level_4_tp_id,
    case when diff <= 0 then '0'
    when diff <= 3 then '0-3'
    when diff <= 7 then '3-7'
    when diff <= 14 then '7-14'
    when diff <= 21 then '14-21'
    when diff <= 30 then '21-30'
    when diff <= 45 then '30-45'
    when diff <= 60 then '45-60'
    when diff <= 90 then '60-90'
    else '>90' end as diff,
    count(distinct a.mobile) as vol
from
(
    select
        a.mobile,
        level_4_tp_id,
        datediff(order_date,last_fir_contact_date_brand) as diff
    from
    (
        select 
            a.mobile,
            c.level_4_tp_id,
            last_fir_contact_date_brand,
            min(action_time) as order_date
        from
        (
            select mobile, last_fir_contact_date_brand, touchpoint_id
            from
            (
                select mobile, last_fir_contact_date_brand, touchpoint_id,
                row_number() over(partition by mobile order by last_fir_contact_date_brand desc) as rank_
                from marketing_modeling.cdm_customer_touchpoints_profile_a
                where brand = "MG"
            ) a
            where rank_ = 1
        ) a
        left join
        (
            select * from marketing_modeling.dw_ts_order_i 
            where brand = 'MG' 
            and touchpoint_id like '011%'
        ) b
        on a.mobile = b.mobile
        left join marketing_modeling.cdm_touchpoints_id_system c
        on a.touchpoint_id = c.touchpoint_id
        where a.last_fir_contact_date_brand <= b.action_time
        group by a.mobile, c.level_4_tp_id, a.last_fir_contact_date_brand
    ) a
) a
group by
    level_4_tp_id,
    case when diff <= 0 then '0'
    when diff <= 3 then '0-3'
    when diff <= 7 then '3-7'
    when diff <= 14 then '7-14'
    when diff <= 21 then '14-21'
    when diff <= 30 then '21-30'
    when diff <= 45 then '30-45'
    when diff <= 60 then '45-60'
    when diff <= 90 then '60-90'
    else '>90' end
'''
```
